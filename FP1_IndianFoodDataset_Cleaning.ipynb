## Cleaning food name and ingredient name for Indian Food dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("./data/inputs/IndianFoodDataset.xlsx")
```

```python
df
```

```python
# Select the desired columns
selected_columns = ['Srno', 'TranslatedRecipeName', 'TranslatedIngredients', 'Cuisine', 'Course', 'Diet']
df_cleaned = df[selected_columns].copy(deep=True)

df_cleaned = df_cleaned.rename(columns={'TranslatedRecipeName': 'recipename', 'TranslatedIngredients': 'ingredients'})

df_cleaned
```

## Clean Recepie names

```python
# Remove the word "Recipe" from the 'recipename' column
df_cleaned['recipename'] = df_cleaned['recipename'].str.replace('Recipe', '')

# remobe () from recipe name
df_cleaned['recipename'] = df_cleaned['recipename'].str.replace('()', '')

# Remove symbols from recipename column
df_cleaned['recipename'] = df_cleaned['recipename'].str.replace(r'[^a-zA-Z\s]', '')

# Remove symbols from the recipename column
df_cleaned['recipename'] = df_cleaned['recipename'].str.replace(r'[-|.]', '').str.replace(r'\W', ' ')

# Remove symbols from the recipename column
df_cleaned['recipename'] = df_cleaned['recipename'].str.replace(r'[-|.]', '')
df_cleaned['recipename']
```

## Clean Ingredients

```python
import pandas as pd
import re

def clean_ingredients(ingredients):
    cleaned_ingredients = []

    # Words and symbols to remove
    remove_words = ['grams', 'gram', 'kg', 'numbers', 'slice', 'stuff', 'thin', 'cup', 'spoon', 'tablespoon', 'inch',
                    'inches', 'chop', 'taste', 'grated', 'as per use', 'pinch', 'teaspoon', 'recipe', 'half', 'full',
                    'quarter', 'above', 'slice', 'thinly', 'thick', 'powder', 'as required', 'cook', 'cooked', 'halved',
                    'soaked', 'tightly']

    remove_symbols = ['-', '/', ',', '.', '*', '#', '@', '$', '%', '^', '&', '(', ')', '[', ']', '{', '}', '|', '<', '>',
                      '=', '+', '_', '`', '~']

    # Remove numbers and fractions
    ingredients = re.sub(r'\b\d+\/\d+\b|\d+\.\d+|\d+', '', str(ingredients))

    for ingredient in ingredients.split(','):
        ingredient = ingredient.strip()

        # Remove text within parentheses
        ingredient = re.sub(r'\(.*?\)', '', ingredient)

        # Remove unwanted words
        for word in remove_words:
            ingredient = ingredient.replace(word, '')

        # Remove unwanted symbols
        for symbol in remove_symbols:
            ingredient = ingredient.replace(symbol, '')

        # Remove additional spaces
        ingredient = ' '.join(ingredient.split())

        # Remove single-letter and two-letter words
        ingredient = ' '.join(word for word in ingredient.split() if len(word) > 2)

        cleaned_ingredients.append(ingredient)

    return cleaned_ingredients


# Clean the "ingredients" column
df_cleaned['ingredients'] = df_cleaned['ingredients'].apply(clean_ingredients)

# Print the updated DataFrame
df_cleaned['ingredients']
```
